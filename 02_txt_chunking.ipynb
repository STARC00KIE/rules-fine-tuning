# 텍스트 청킹
1. '장' 별로 청킹 진행 및 저장

```python
import re
```

```python
# 파일 읽기
input_file = "./data/full_output_fix_2.txt"

with open(input_file, "r", encoding="utf-8") as f:
    text = f.read()

print(text)
```

```python
# 챕터 리스트
chapters = ["1. 총칙", "2. 고용", "3. 복무", "4. 휴가제도", "5. 보수 및 보상", "6. 사회보장 제도", "7. 퇴직 규정", "8. 포상 및 징계", "9. 남녀평등", "10. 직장 내 괴롭힘의 예방", "11. 성과평가", "12. 인사기록", "13. 보칙"]
```

```python
# 리스트를 이용해 정규식 패턴 만들기
pattern = "(" + "|".join(map(re.escape, chapters)) + ")"

# 텍스트 자르기
parts = re.split(pattern, text)

# 결과를 딕셔너리에 저장
chunked_data = {}

# 인덱스 1부터 시작해서 2칸씩 건너뛰며 제목과 내용을 짝지음
for i in range(1, len(parts), 2):
    title = parts[i].strip()       # 챕터 제목
    content = parts[i+1].strip()   # 챕터 내용
    chunked_data[title] = content

# 결과 확인
print(f"총 {len(chunked_data)}개의 챕터 추출\n")

for title, content in chunked_data.items():
    print(f"[{title}] : {len(content)} 글자")
    # 내용 앞부분 50자만 미리보기 (필요하면 주석 해제)
    print(f"   내용 미리보기: {content[:50]}...")
    print("-" * 30)

    # 제목 변경
    safe_title = title.replace(". ", "_")
    file_name = f"./data/chunking_raw/{safe_title}.txt"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write(content)
```

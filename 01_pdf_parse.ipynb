```python
# 임포트
import fitz
import re
```

```python
path = "./docs/취업규칙_브이티더블유_20230829.pdf"
raw_text = ""
doc = fitz.open(path)
```

```python
for page in doc[3:]:
    text = page.get_text()
    raw_text += text
    print(text)
    
```

```python
text = raw_text[5:]
text = text.replace("ž", "·")

# 1. 특수문자 치환
text = text.replace("ž", "·")

# 2. [초기] 불필요한 페이지 번호 제거
text = re.sub(r'\n\s*\d+\s*\n', '\n', text) 

# 3. 공백 정규화
text = re.sub(r'\s+', ' ', text)

# 3-1. [오타 교정] 글자와 조사/어미 사이 공백 제거
text = re.sub(r'([가-힣])\s+([은는이가을를에의하])', r'\1\2', text)
text = re.sub(r'(아니하)\s+(다)', r'\1\2', text) # '아니하 다' 처리

# 3-2. [괄호 공백 교정] 숫자와 닫는 괄호 사이 공백 제거
text = re.sub(r'(\d+)\s+\)', r'\1)', text)

# 4. [문단 번호] 구조 복원 (1. / 1.1. / 2.2.1 등)
regex_chapter = r'(\s)((?:\d+\.)+(?:\d+)?)\s' 
text = re.sub(regex_chapter, r'\n\n\2 ', text)

# 5. [괄호 번호] 복원 (1), 2) 등)
text = re.sub(r'(\s)(\d+\))', r'\n\2', text)

# 6. [원문자] 복원 (①, ② 등)
text = re.sub(r'([①-⑮])', r'\n\1', text)

# 7. [알파벳 리스트] 복원 (A. B. C.)
text = re.sub(r'(\s)([A-Z]\.\s)', r'\n\2', text)

# 8. [후처리] 문장 끝에 남은 페이지 번호 제거
text = re.sub(r'(?<=\.)\s+\d+\s*(?=\n)', '', text)


cleaned_text = text.strip()
print(cleaned_text)
```

```python
# 텍스트 저장
f = open("./data/full_output.txt", "w", encoding="utf-8")
f.write(cleaned_text)
f.close()
```
